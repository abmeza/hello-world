```python
import pandas as pd

data = pd.read_csv('number-of-internet-users-by-country.csv')

#drop Code column
data = data.drop(labels='Code', axis=1)

# Get info about users in 2010 specifically
Afghanistan_info = data[data.Entity == 'Afghanistan']
Afghanistan_info = Afghanistan_info[ Afghanistan_info.Year == 2010]
Albania_info = data[data.Entity == 'Albania']
Albania_info = Albania_info[ Albania_info.Year == 2010]
Algeria_info = data[data.Entity == 'Algeria']
Algeria_info = Algeria_info[ Algeria_info.Year == 2010]
Angola_info = data[data.Entity == 'Angola']
Angola_info = Angola_info[ Angola_info.Year == 2010]

Antigua_info = data[data.Entity == 'Antigua and Barbuda']
Antigua_info = Antigua_info[ Antigua_info.Year == 2010]
Argentina_info = data[data.Entity == 'Argentina']
Argentina_info = Argentina_info[ Argentina_info.Year == 2010]

labels = ['Afghanistan', 'Albania', 'Algeria', 
          'Angola', 'Antigua', 'Argentina']
values = []

all_data= pd.concat([Afghanistan_info, 
                        Albania_info,
                        Algeria_info,
                        Angola_info,
                        Antigua_info,
                        Argentina_info]) 

print(all_data)
```

```python
import matplotlib.pyplot as plt
all_data.plot(x="Entity", y="Number of internet users (OWID based on WB & UN)", kind="bar")

```
